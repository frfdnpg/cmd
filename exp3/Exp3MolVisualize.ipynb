This notebooks aims at visualizing the structures of clusters of different sets of natural products

```python
# Import myfunc at cix folder
%matplotlib inline
import matplotlib.pyplot as plt
import sys
sys.path.insert(0, '../cix')
import myfuncs as mf
import pandas as pd
from rdkit import Chem
from rdkit import rdBase
rdBase.DisableLog('rdApp.*') # To make rdkit silent
from rdkit.Chem import PandasTools as pt
from rdkit.Chem import Descriptors
import numpy as np
import time
```

```python
##############################################################################
## Input molecules
##############################################################################

d = pd.read_csv("./nppurg2.csv")
smis = list(d["smi"])
ncorr, n, smis, wrongsmis = mf.corrsmis(smis)
smidf = mf.smis2smidf(smis)
cls = mf.clusmidf(smidf)
```

```python
cl_smis_in = []
for i in range(10):
    cl = cls[i]
    smis = [smidf.loc[smidf["id"] == i, "smiles"].iloc[0] for i in cl]
    cl_smis_in.append(smis)
```

```python
mf.paintmols(cl_smis_in[0])
```
